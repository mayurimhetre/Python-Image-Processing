```python
import numpy as np 
import matplotlib.pyplot as plt
import cv2
```

```python
a = cv2.imread('00-puppy.jpg')
```

```python
cv2.imshow('names',a)
```

```python
while True:
    cv2.imshow('names',a)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cv2.destroyAllWindows()
```

```python
plt.imshow(a)
```

```python
fix_img  = cv2.cvtColor(a,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(fix_img)
```

```python
################# plotting polygons 
fix_img1  = cv2.cvtColor(a,cv2.COLOR_BGR2RGB)
arr = np.array([[750,1000],[1000,400],[1250,400],[1300,600],[1500,800]],dtype=np.int32)
arr =arr.reshape(-1,1,2)
img2 = cv2.polylines(fix_img1,[arr],isClosed=True,color= (0,255,255),thickness= 10)
plt.imshow(img2)
```
